```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## L1-Scan

```python
df1 = pd.read_csv("L1_scan/L1_1_5.txt", skiprows=1)
df2 = pd.read_csv("L1_scan/L1_0_5.txt", skiprows=1)

# trim off whitespace from column names
df1.columns = [col.strip() for col in df1.columns]
df2.columns = [col.strip() for col in df2.columns]

# Grab Λ and N
lams1 = np.unique(np.array(df1["LAMBDA"],dtype=float))
lams2 = np.unique(np.array(df2["LAMBDA"],dtype=float))
Ns1 = np.unique(np.array(df1["N"],dtype=float))
Ns2 = np.unique(np.array(df2["N"],dtype=float))

# drop nans
lams1 = lams1[np.logical_not(np.isnan(lams1))]
lams2 = lams2[np.logical_not(np.isnan(lams2))]
Ns1 = Ns1[np.logical_not(np.isnan(Ns1))]
Ns2 = Ns2[np.logical_not(np.isnan(Ns2))]

# grab Ω's
rd_es1 = np.array(df1["RD_ETA"],dtype=float).reshape((len(lams1),len(Ns1)))
rd_es2 = np.array(df2["RD_ETA"],dtype=float).reshape((len(lams2),len(Ns2)))

rd_ds1 = np.array(df1["RD_DELTA"],dtype=float).reshape((len(lams1),len(Ns1)))
rd_ds2 = np.array(df2["RD_DELTA"],dtype=float).reshape((len(lams2),len(Ns2)))

# grab ΔN_effs
delta_Neff_cmb1 = np.nan_to_num(np.array(df1["DELTA_NEFF_CMB"],dtype=float).reshape((len(lams1),len(Ns1))))
delta_Neff_bbn1 = np.nan_to_num(np.array(df1["DELTA_NEFF_BBN"],dtype=float).reshape((len(lams1),len(Ns1))))

delta_Neff_cmb2 = np.nan_to_num(np.array(df2["DELTA_NEFF_CMB"],dtype=float).reshape((len(lams2),len(Ns2))))
delta_Neff_bbn2 = np.nan_to_num(np.array(df2["DELTA_NEFF_BBN"],dtype=float).reshape((len(lams2),len(Ns2))))

# grab si cs
si_e1 = np.nan_to_num(np.array(df1["(SIG_ETA / M_ETA)"],dtype=float).reshape((len(lams1),len(Ns1))))
si_d1 = np.nan_to_num(np.array(df1["(SIG_DELTA / M_DELTA)"],dtype=float).reshape((len(lams1),len(Ns1))))

si_e2 = np.nan_to_num(np.array(df2["(SIG_ETA / M_ETA)"],dtype=float).reshape((len(lams2),len(Ns2))))
si_d2 = np.nan_to_num(np.array(df2["(SIG_DELTA / M_DELTA)"],dtype=float).reshape((len(lams2),len(Ns2))))

# create averages weighted by RD's for self interaction cross-sections
si1 = (si_e1 * rd_es1**2 + si_d1 * rd_ds1**2) / (rd_es1**2 + rd_ds1**2)
si2 = (si_e2 * rd_es2**2 + si_d2 * rd_ds2**2) / (rd_es2**2 + rd_ds2**2)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(7,4))
f.dpi=150

ax1.contour(Ns1, lams1, rd_es1, levels=[0.112], colors='firebrick')
ax1.contour(Ns1, lams1, rd_ds1, levels=[0.112], colors='steelblue')
ax1.contour(Ns1, lams1, rd_es1+rd_ds1, levels=[0.112])
ax1.contour(Ns1, lams1, si1, levels=[2.19893e14],colors='goldenrod')

ax1.set_yscale('log')
ax1.set_xscale('log')
ax1.set_ylabel(r'$\Lambda \ (\mathrm{GeV})$', fontsize=16)
ax1.set_xlabel(r'$N$', fontsize=16)
ax1.set_title(r'$L_{1} = 3/2$', fontsize=16)

ax2.contour(Ns2, lams2, rd_es2, levels=[0.112], colors='firebrick')
ax2.contour(Ns2, lams2, rd_ds2, levels=[0.112], colors='steelblue')
ax2.contour(Ns2, lams2, rd_es2 + rd_ds2, levels=[0.112])
ax2.contour(Ns2, lams2, si2, levels=[2.19893e14],colors='goldenrod')
ax2.set_yscale('log')
ax2.set_xscale('log')
ax2.set_xlabel(r'$N$', fontsize=16)
ax2.set_title(r'$L_{1} = 1/2$', fontsize=16)

f.tight_layout()
```
