```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
data_zs = np.genfromtxt("log_scaled_cs_data_zs.dat")
data_cs44 = np.genfromtxt("log_scaled_cs_data_cs44.dat")
data_cs66 = np.genfromtxt("log_scaled_cs_data_cs66.dat")
data_cs46 = np.genfromtxt("log_scaled_cs_data_cs46.dat")
```

```python
plt.figure(dpi=150)
plt.plot(data_zs,data_cs44)
plt.plot(data_zs,data_cs66)
plt.plot(data_zs, data_cs46)
```

```python
def model(x, b):
    return 14 * x + b

popt44, _ = curve_fit(model, data_zs[-50:], data_cs44[-50:])
popt66, _ = curve_fit(model, data_zs[-50:], data_cs66[-50:])
popt46, _ = curve_fit(model, data_zs[-50:], data_cs46[-50:])
```

```python
print(popt44)
print(popt66)
print(popt46)
```
